```python
likesql1 = pd.DataFrame({'key': ['k1', 'k2', 'k3'],
                        'data1': [1,2,3],
                        'data2': [4,5,6]})
```

```python
likesql2 = pd.DataFrame({'key': ['k1', 'k2', 'k3'],
                        'data1': [1,2,3],
                        'data2': [4,5,6]})
```

```python
likesql1
```

### Merge

```python
likesqljoin = pd.merge(likesql1, likesql2, how = 'inner', on='key')
```

```python
likesqljoin
```

merge ini seperti join di SQL, jadi ada inner, left dll, lalu untuk key joinnya bisa juga 1 key atau list (beberapa key)

Ada juga fungsi join, syaratnya keynya harus index dari dataframenya

```python
test1 = pd.DataFrame({'A': ['wow', 'wew'],
                     'B': ['wiw', 'wuw']},
                    index=['index1', 'index2'])
```

```python
test2 = pd.DataFrame({'C': ['pomp', 'pemp'],
                     'D': ['pimp', 'pump']},
                    index=['index1', 'index3'])
```

```python
test1
```

```python
test2
```

```python
test1.join(test2)
```

```python
test1.join(test2, how='inner')
```

```python
test1.join(test2, how='left')
```

```python
test1.join(test2, how='right')
```

```python
test1.join(test2, how='outer')
```
